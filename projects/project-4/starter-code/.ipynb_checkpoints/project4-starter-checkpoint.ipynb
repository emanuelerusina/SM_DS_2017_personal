# Project 4

In this project, you will summarize and present your analysis from Projects 1-3.

### Intro: Write a problem Statement/ Specific Aim for this project

Answer: The goal of this project is to explore the UCLA admission data to identify a potential association between GRE score, GPA, Undergrad Alma Mater prestige and whether the candidate was admitted to UCLA.

### Dataset:  Write up a description of your data and any cleaning that was completed

Answer: The data provided is relative to 400 candidates, and includes their Grade Point Average (2.8-4.0), GRE score (260-800) and Undergrad Alma Mater prestige (1-4, in descending prestige). All of the record with any missing observation has been removed from the sample, resulting in 387 observations.

### Methods: Write up the methods used in your analysis

Answer:The analysis performs a Logistic Regression on the dataset with Admission as dependent variable and GRE score, GPA and Prestige as covariates. 

### Results: Write up your results

Answer:The analysis found a significant relationship between prestige of the undergrad alma mater prestige (quantified on a four-point scale), controlling for both GPA and GRE score. Graduating from the top category alma maters effectively doubles the odds of admission relative to the second layer in prestige.

### Discussion: Write up your  and future steps

Answer: The dataset is missing time frame information, so a potential follow-up could consider exploring changes to the odds over time. 
Supplementing the dataset with additional demographic/geographic information may also provide insights about relevant misalignments in the characteristics of propesctive students within different Alma Mater ranks, as well as reveal potential unconscious bias in the admission committee's decisions.
